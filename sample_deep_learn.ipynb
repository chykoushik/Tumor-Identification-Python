```python
import pandas as pd
import numpy as np
import tensorflow as tf

seed = 2309
np.random.seed(seed)
```

```python
dat = pd.read_csv("norm_data__non_log.txt",sep='\t')
label = pd.read_csv("sample_list.csv",sep=';')
```

```python
real_dat = dat.T
```

```python
real_dat.shape
```

```python
mapping = {'Non-LCa':0,'LCa':1}
target = label.Disease.map(mapping)
```

```python
## Feature Selection
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, mutual_info_classif

X_new = SelectKBest(chi2, k=30).fit_transform(real_dat, target)
X_new.shape
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

A = np.asmatrix(real_dat.values) * np.asmatrix(real_dat.T.values)
U, S, V = np.linalg.svd(A) 
eigvals = S**2 / np.sum(S**2)  # NOTE (@amoeba): These are not PCA eigenvalues. 
                               # This question is about SVD.

fig = plt.figure(figsize=(8,5))
sing_vals = np.arange(real_dat.shape[1]) + 1
plt.plot(sing_vals, eigvals, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
#I don't like the default legend so I typically make mine like below, e.g.
#with smaller fonts and a bit transparent so I do not cover up data, and make
#it moveable by the viewer in case upper-right is a bad place for it 
leg = plt.legend(['Eigenvalues from SVD'], loc='best', borderpad=0.3, 
                 shadow=False, prop=matplotlib.font_manager.FontProperties(size='small'),
                 markerscale=0.4)
leg.get_frame().set_alpha(0.4)
leg.draggable(state=True)
plt.show()
```
