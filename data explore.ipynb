```python
import os
import re 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Data
data = pd.read_csv("norm_data__non_log.txt",sep='\t')
data.sample(n=3, random_state=1)
```

```python
# Check data null
sum(data.isnull().sum())
sum(data.isna().sum())
```

```python
# Check feature distribution

features = {'mean': data.mean(1),
        'sd': data.std(1)
        }

df = pd.DataFrame(features, columns= ['mean', 'sd'])
df
```

```python
# Extract labels
def extract_label(list): 
    number = '[0-9]'
    symbol = '_'
    head = 'Sample'
    list = [re.sub(number, '', i) for i in list] 
    list = [re.sub(symbol, '', i) for i in list] 
    list = [re.sub(head, '', i) for i in list] 
    return list

labels = list(data)
labels = extract_label(labels)
labels = np.ravel(labels)


data=data.T

mapping = {'Non-LCa':-1,'LCa':1}
data["targets"] = labels
data["targets"] = data["targets"].map(mapping)
#data
data.sample(n=3)


# Check label distribution

#a = labels.count('Non-LCa')
#b = labels.count('LCa')
#print('Non-LCa count:',a,'\tLCa count:',b)
#print('Non-LCa ratio:',a/(a+b),'\tLCa ratio:',b/(a+b))


```
